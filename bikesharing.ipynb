```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:/Users/User/Downloads/BikeSharingData.csv")
df.drop(['instant'],axis=1)
df.tail()
```

```python
plt.subplot(121)
plt.xlabel('casual')
plt.scatter(df['casual'],df['cnt'])
plt.subplot(122)
plt.xlabel('registered')
plt.scatter(df['registered'],df['cnt'])
```

```python
# g = sns.FacetGrid(df, col='yr')
# g = g.map(plt.hist,"cnt", color="r")
# plt.show()
plt.bar(df['yr'],df['cnt'])
```

```python
g = sns.FacetGrid(df, col='season')
g = g.map(plt.hist,"cnt", color="r")
plt.show()
```

```python
g = sns.FacetGrid(df, col='weathersit')
g = g.map(plt.hist,"cnt", color="y")
plt.show()
```

```python
g = sns.FacetGrid(df, col='holiday')
g = g.map(plt.hist,"cnt", color="r")
plt.show()
```

```python
g = sns.FacetGrid(df, col='weekday')
g = g.map(plt.hist,"cnt", color="r")
plt.show()
```
