```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
a=list(['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','country','salary'])
data=pd.read_csv("C:/Users/User/Desktop/adultdata.csv",names=a,na_values='?')
data
# pd.get_dummies(data)
```

```python
data.fillna(np.NaN)
data.isnull().sum()
# salary_map={'<=50K':1,'>50K':0}
# data['salary']=data['salary'].map(salary_map).astype(int)
data['salary']=data['salary'].replace(' <=50K', 0)
data['salary']=data['salary'].replace(' >50K', 1)

data.head(10)
```

```python
data['workclass'].value_counts()
```

```python
data.replace([' Without-pay',' Never-worked'],['unemployed','unemployed'],inplace=True)
import seaborn as sns
sns.factorplot(x="workclass", y="salary", data=data, kind="bar", size = 6, 
palette = "muted")
plt.xticks(rotation=45);
```

```python
data.replace([' Divorced', ' Married-AF-spouse', 
              ' Married-civ-spouse', ' Married-spouse-absent', 
              ' Never-married',' Separated',' Widowed'],
             ['not married','married','married','married',
              'not married','not married','not married'], inplace = True)
data.head()
```

```python
sns.factorplot(x="education",y="salary",data=data,kind="bar", size = 7, 
palette = "muted")
plt.xticks(rotation=60);
```

```python
plt.scatter(data['age'],data['salary'],data=data)
plt.show()
```

```python
category_col =['workclass', 'race', 'education','marital-status', 'occupation',
               'relationship', 'sex', 'country'] 
for col in category_col:
    b, c = np.unique(data[col], return_inverse=True) 
    data[col] = c
data
```

```python
f, ax = plt.subplots(figsize=(30, 10))
corr = data.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(240,10,as_cmap=True),
            square=True, ax=ax,annot=True)
```
